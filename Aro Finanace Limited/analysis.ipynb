```python
def count_lower(input_list):
    result = [0] * len(input_list)
    for i in range(0,len(input_list)):
        inner_list = input_list[i+1:]
        count = 0
        for j in inner_list:
            if input_list[i] > j:
                count+= 1
        result[i] = count
    return result

print(count_lower([5,6,3,2,1]))
```

```python
def sum_of_list(input_list):
    data_count = {}
    sum = 0
    for i in input_list:
        data_count[i] = 0
    for i in input_list:
        if data_count[i] == 1:
            data_count[i] = data_count[i]+1
        elif data_count[i] == 0:
            data_count[i] = 1
    for key, value in data_count.items():
        if value == 1:
            sum = sum+key
    return sum

print(sum_of_list([1,10,3,10,10]))
```
