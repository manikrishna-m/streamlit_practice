```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("pokemon_data.csv")
dataset
```

```python
plt.figure(figsize =(16,8))
plt.stem(dataset["HP"])
plt.title("HP vs Attack")
plt.xlabel("HP")
plt.ylabel("Attack")
plt.show()
```

```python
plt.figure(figsize =(16,8))
plt.style.use('fivethirtyeight')
plt.plot(dataset["#"], dataset["Attack"],color='k', linestyle='--',marker="o", label = "Attack", linewidth = 3)
plt.plot(dataset["#"], dataset["HP"],color='b',linestyle = "-",marker="o", label = "HP")
plt.title("HP vs Attack")
plt.xlabel("HP")
plt.ylabel("Attack")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

np.random.seed(1)
x = 4 + np.random.normal(0, 1.5, 200)

plt.figure(figsize = (8,8))

plt.hist(x, bins=8, linewidth=0.5, edgecolor="white")

plt.xlim(0,8)
plt.ylim(0,56)
plt.xticks(np.arange(1, 8))
plt.yticks(np.linspace(0, 56, 9))
plt.title("Histogram")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
def planidrome(number):
    return str(number) == str(number)[::-1]

planidrome(121)
```

```python
planidrome(1212)
```

```python
def fact(n):
    if n==0 or n ==1:
        return 1
    else:
        return n * fact(n-1)

fact(3)
```

```python
fact(4)
```

```python
fact(5)
```

```python
def sort_list(list1, list2):
    return sorted(list1+list2)
```
