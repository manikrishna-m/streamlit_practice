```python
import pandas as pd

raffle_mailings_df = pd.read_csv('../data/raffle_mailings_sample_data.csv')

print("Raffle Mailings Data Info")
display(raffle_mailings_df.info())

print("Raffle Mailings Data")
display(raffle_mailings_df.head())

raffle_mailings_df["LotteryPlays"].fillna(0, inplace = True)
raffle_mailings_df["RecruitmentSource"].fillna(raffle_mailings_df["RecruitmentSource"].mode()[0], inplace = True)
raffle_mailings_df.loc[(raffle_mailings_df["Title"].isna()) & (~raffle_mailings_df["Gender"].isna()), "Title"] = "Mr"
raffle_mailings_df["Gender"].fillna(raffle_mailings_df["Gender"].mode()[0], inplace = True)
raffle_mailings_df["Title"].fillna(raffle_mailings_df["Title"].mode()[0], inplace = True)
raffle_mailings_df["RaffleMailingDate"] = pd.to_datetime(raffle_mailings_df["RaffleMailingDate"])

print("Raffle Mailings Data Info After data cleaning and data correction")
display(raffle_mailings_df.info())
```

```python
raffle_entries_df = pd.read_csv('../data/raffle_entries_sampled_data.csv')

print("Raffle Entries Dataframe Info:")
display(raffle_entries_df.info())

print("Raffle Entries Data")
display(raffle_entries_df.head())

print("Duplicates in Raffle Entries Data:", raffle_entries_df.duplicated().sum())
raffle_entries_df.drop_duplicates(inplace= True)

raffle_entries_df["GiftAidAmount"].fillna(0, inplace = True)
raffle_entries_df["DateEnteredRaffle"] = pd.to_datetime(raffle_entries_df["DateEnteredRaffle"])

print("Raffle Entries Data Info After data cleaning and data correction, dupliacted")
display(raffle_entries_df.info())
```

```python
merged_df = raffle_mailings_df.merge(raffle_entries_df, how="left", on="UniqueId")

merged_df['PaymentType'].fillna('None', inplace=True)
merged_df['DonationAmount'].fillna(0, inplace=True)
merged_df['GiftAidAmount'].fillna(0, inplace=True)
merged_df['DateEnteredRaffle'].fillna('1970-01-01 00:00:00', inplace=True)

merged_df['EnteredRaffle'] = merged_df['DateEnteredRaffle'].apply(lambda x: 0 if x == pd.to_datetime('1970-01-01 00:00:00') else 1)

merged_df =  merged_df[['RaffleMailingDate', 'Title', 'Gender', 'RecruitmentSource','LotteryPlays', 'UniqueId','EnteredRaffle']]

print("Raffle Entries Data with label")
merged_df.info()
```

```python
print("Duplicates in Raffle Entries Data with label:", merged_df.duplicated().sum())

merged_df.drop_duplicates(keep = "last", inplace = True)
```
